```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
df = pd.read_csv('examdata.csv')
print(df)
```

```python
X = df.drop(columns=['Pass/Fail'])
Y = df['Pass/Fail']
X
#Y
```

```python
Y
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(X,Y)
```

```python
ans = model.predict([[4.370861,81.889703]])
print(ans)
```

```python
df1 = model.score(X,Y)
print(df1)
```
